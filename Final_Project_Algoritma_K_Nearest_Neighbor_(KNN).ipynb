#**Tabel confusion matrix**

Confusion matrix dapat digunakan untuk mengukur performa dalam permasalahan klasifikasi biner maupun permasalahan klasifikasi multiclass. **Contoh Penerapan Confusion Matrix**


```python
import pandas as pd

data = {'y_Actual':    [1, 0, 0, 1, 0, 1, 0, 0, 1, 0, 1, 0],
        'y_Predicted': [1, 1, 0, 1, 0, 1, 1, 0, 1, 0, 0, 0]
        }

df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])
print (df)
```

```python
confusion_matrix = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])
print (confusion_matrix)
```

**Menampilkan confusion matrix menggunakan seaborn**

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

data = {'y_Actual':    [1, 0, 0, 1, 0, 1, 0, 0, 1, 0, 1, 0],
        'y_Predicted': [1, 1, 0, 1, 0, 1, 1, 0, 1, 0, 0, 0]
        }

df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])
confusion_matrix = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])

sn.heatmap(confusion_matrix, annot=True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Untuk memahami konsep, kita akan menggunakan data X dan Y seperti yang kita tunjukkan pada proses penghitungan manual

```python
X=np.array([[1,1], [2,2], [3,2], [12,11], [11,13], [14,15]])
Y=np.array([0,0,0,1,1,1])
x_baru=np.array([9,10])
```

```python
X
```

Menentukan parameter k (jumlah tetangga paling dekat)

```python
k=3
```

**Langkah 2:**

Menghitung kuadrat jarak eucliden objek terhadap data training yang diberikan.

```python
def jarak_euclidean(x,y):
    return np.sqrt(sum(pow(a-b, 2) for a, b in zip(x,y)))
```

```python
jarak=[]
for x_i in X:
    jarak_euclid=jarak_euclidean(x_i, x_baru)
    jarak.append(jarak_euclid)
    print(jarak)
print(jarak)
```

**Langkah 3:**

Mengurutkan hasil langkah 2 secara ascending (berurutan dari rendah ke tinggi)

```python
urut=np.sort(jarak)
list_urut=[]
for i in range(len(urut)):
    list_urut.append(urut[i])
print(list_urut)
```

```python
urutan=[]
for i in list_urut:
    urtn=jarak.index(i)
    urutan.append(urtn)
print(urutan)
```

Kelas data 3 data-data terdekat (k=3)

## **KNN dengan Jarak Minkowski**

Jarak Minkowski atau Metrik Minkowski adalah metrik di sebuah ruang vektor bernorma yang dapat dianggap sebagai generalisasi dari keduanya Jarak Euclidean dan Jarak Manhattan.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=np.array([[1,1], [2,2], [3,2], [12,11], [11,13], [14,15]])
Y=np.array([0,0,0,1,1,1])
x_baru=np.array([9,10])
```

```python
X
```

**Langkah 1:**

```python
k=3
```

**Langkah 2:**

Menghitung kuadrat jarak minkowski objek terhadap data training yang diberikan.

```python
def akar_ke_n(nilai, akar_n):
  nilai_akar=1/float(akar_n)
  return round(float(nilai)**float(nilai_akar),3)

def minkowski(x,y,p_value):
  return akar_ke_n(sum(pow(abs(a-b), p_value) for a, b in zip(x,y)), p_value)
```

```python
jarak=[]
for x_i in X:
    jarak_minkowski=minkowski(x_i, x_baru,1)
    jarak.append(jarak_minkowski)
    print(jarak)
print(jarak)
```

Jarak Chebyshev merupakan metode yang mengukur jarak berdasarkan nilai mutlak atau absolut dari selisih pasangan koordinat sebuah titik.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=np.array([[1,1], [2,2], [3,2], [12,11], [11,13], [14,15]])
Y=np.array([0,0,0,1,1,1])
x_baru=np.array([9,10])
```

**Langkah 2:**

Menghitung kuadrat jarak Bray-Curtiz objek terhadap data training yang diberikan.

```python
def jarak1(x,y): # Perhitungan Jarak Bray_Curtiz
    return sum(abs(a-b) for a,b in zip(x,y))

def jarak2(x,y): # Perhitungan Jarak Bray_Curtiz
    return sum(abs(a+b) for a,b in zip(x,y))
```

```python
jarak=[]
for x_i in X:
    jarak_BC1=jarak1(x_i, x_baru)
    jarak_BC2=jarak2(x_i, x_baru)
    jarak_BC = jarak_BC1 / jarak_BC2
    jarak.append(jarak_BC)
    print(jarak)
print(jarak)
```

### Plot Function

```python
def bar_plot(df, evalf):

    # Memilih tema plot
    sns.set_theme(style="whitegrid")

    # Membuat sebuah figure dan sebuah axes
    fig, ax=plt.subplots(figsize=(10,8)) # figsize=(x,y) untuk menentukan ukuran gambar

    # Melakukan bar plot
    ax = sns.barplot(data=df,               # Nama DataFrame
                     y=evalf,               # Data sumbu Y
                     x=df.index             # Data sumbu X

                    )
    for bar in ax.patches:
        ax.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')
    # Mengatur sudut rotasi label stepsize sumbu x
    plt.xticks(rotation=80)

    # Mengganti label sumbu x
    ax.set_xlabel('distance')

    # Mengganti label sumbu y
    ax.set_ylabel('{} score'.format(evalf))

    # Mengatur Judul Gambar
    ax.set_title('Evaluation Score of KNN with Various Distance', fontsize=16, fontweight='bold')

    # Menampilkan Gambar
    plt.show()
```
